```python


# Quadratic Functions
import matplotlib.pyplot as plt
import numpy as np

# Define the quadratic functions
def f1(x):
    return -x**2 + x - 2

def f2(x):
    return x**2 + 3*x - 2

# Generate x values
x = np.linspace(-100, 100, 400)

# Generate y values
y1 = f1(x)
y2 = f2(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = 2x^2 + 3x + 3')
plt.plot(x, y2, label='y = -2x^2 + 3x + 3')

# Add title and labels
plt.title('Quadratic Functions')
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import math

# Fungsi untuk mencari akar persamaan kuadrat
def cariakar(a, b, c):
    # Hitung diskriminan
    D = b**2 - 4*a*c
    
    # Gunakan if-else untuk menentukan jenis akar
    if D > 0:
        root1 = (-b + math.sqrt(D)) / (2*a)
        root2 = (-b - math.sqrt(D)) / (2*a)
        return f"Dua akar nyata: {root1} dan {root2}"
    elif D == 0:
        root = -b / (2*a)
        return f"Satu akar nyata (akar kembar): {root}"
    else:
        real_part = -b / (2*a)
        imaginary_part = math.sqrt(-D) / (2*a)
        return f"Dua akar imajiner: {real_part} + {imaginary_part}i dan {real_part} - {imaginary_part}i"

# Contoh penggunaan
a = 1
b = -3
c = 2

result = cariakar(a, b, c)
print(result)

a = 1
b = 2
c = 1

result = cariakar(a, b, c)
print(result)

a = 1
b = 1
c = 1

result = cariakar(a, b, c)
print(result)
```

```python
# NAMA : FADHIL MUDDASIR
# NIM : 24723301
# Function Fibonacci
import numpy as np

def iter_fib(n):
    fib = np.ones(n)
    
    for i in range(2, n):
        fib[i] = fib[i - 1] + fib[i - 2]
        
    return fib    

%timeit iter_fib(25)
```

```python
def fibonacci(n):
    """Computes and returns the Fibonacci of n, 
    a postive integer.
    """
    if n == 1: # first base case
        return 1
    elif n == 2: # second base case
        return 1
    else: # Recursive step
        return fibonacci(n-1) + fibonacci(n-2) # Recursive call 

%timeit fibonacci(25)
```
