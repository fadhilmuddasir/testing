```python
import random

angka_rahasia = random.randint(1, 5)
#print(angka_rahasia)

tebakan = None

while tebakan != angka_rahasia:
    tebakan = int(input("Tebak angka antara 1 dan 5: "))
    if tebakan < angka_rahasia:
        print("Terlalu rendah!")
    elif tebakan > angka_rahasia:
        print("Terlalu tinggi!")
    else:
        print("Selamat, Anda benar!")
```

```python
import random
import matplotlib.pyplot as plt

f = open('datarand.txt', 'w')

# Menghasilkan 5 bilangan acak antara 0 dan 10
x_values = [random.uniform(0, 10) for _ in range(5)]
# Menghitung f(x) = x^2
y_values = [x**2 for x in x_values]

f.write(f"x={x_values}, y={y_values}\n")
f.close()
# Membuat scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, color='b', alpha=0.6)

# Menambahkan judul dan label
plt.title('Plot f(x) = x^2 dengan x Bilangan Acak')
plt.xlabel('x (Bilangan Acak)')
plt.ylabel('f(x)')

# Menampilkan plot
plt.grid()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Menghasilkan 10 bilangan acak antara 1 hingga 5
angka_acak = [random.randint(1, 5) for _ in range(10)]

# Membuat plot
plt.figure(figsize=(10, 6))
plt.hist(angka_acak, bins=range(1, 52), edgecolor='black', alpha=0.7)

# Menambahkan judul dan label
plt.title('Histogram Bilangan Acak dari 1 hingga 5')
plt.xlabel('Bilangan')
plt.ylabel('Frekuensi')

# Menampilkan plot
plt.xticks(range(1, 6))
plt.grid(axis='y')
plt.show()
```
